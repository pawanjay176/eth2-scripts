```python
import pandas as pd
import remerkleable
import numpy as np
```

```python
df_attestations = pd.read_csv("data/attestations.csv")
df_attestations.fillna("", inplace = True)

df_aggregates = pd.read_csv("data/aggregates.csv")
df_aggregates.fillna("", inplace = True)
```

```python
def get_peer_to_clients(df):
    d1 = dict(zip(df_attestations["propagation_source"], df_attestations["propagation_client"]))
    d1 = dict(filter(lambda x: x[1]!='Unknown', d1.items()))
    d2 = dict(zip(df_attestations.message_source, df_attestations.message_client))
    d2 = dict(filter(lambda x: x[0]!="" and (x[1]!="" and x[1] != 'Unknown'), d2.items()))
    
    for (key, value) in d2.items():
        if key not in d1:
            d1[key] = value
    return d1
```

```python
d1 = get_peer_to_clients(df_attestations)
d2 = get_peer_to_clients(df_aggregates)
peer_to_clients = d1
for (k, v) in d2.items():
    if k not in d1:
        peer_to_clients[k] = v
```

```python
df_attestations['propagation_client'] = df_attestations.propagation_source.apply(lambda x: peer_to_clients[x] if x in peer_to_clients else "Unknown")

print(df_attestations['message_client'].value_counts())

df_attestations['message_client'] = df_attestations.message_source.apply(lambda x: peer_to_clients[x] if x!="" and x in peer_to_clients else "Unknown")

print(df_attestations['message_client'].value_counts())
```

```python
df_aggregates['propagation_client'] = df_aggregates.propagation_source.apply(lambda x: peer_to_clients[x] if x in peer_to_clients else "Unknown")

print(df_aggregates['message_client'].value_counts())

df_aggregates['message_client'] = df_aggregates.message_source.apply(lambda x: peer_to_clients[x] if x!="" and x in peer_to_clients else "Unknown")

print(df_aggregates['message_client'].value_counts())
```
